```python
#import libraris
from bs4 import BeautifulSoup
import requests
from csv import writer
import csv
```

```python
# read url
url="https://dothi.net/nha-dat-ban-ha-noi.htm"
page=requests.get(url)
print(page)
```

```python
# use bs to read content
soup=BeautifulSoup(page.content,'html.parser')
lists=soup.find_all('div',class_="desc")
print(soup.prettify())
```

```python
# list content needed
for listings in lists:
    price = listings.find('div',class_="price").text.replace('\r\n','').replace('\n','')
    area = listings.find('div',class_="area").text.replace('\r\n','').replace('\n','')
    location = listings.find('div',class_="location").text.replace('\r\n','').replace('\n','')
    info=[price,area,location]
    print(info)
```

```python
# save in csv file
with open('housing.csv','w',encoding='utf-8-sig',newline='') as f:
    thewriter=writer(f)
    header=['price','area','location']
    thewriter.writerow(header)
    for listings in lists:
        price = listings.find('div',class_="price").text.replace('\r\n','').replace('\n','')
        area = listings.find('div',class_="area").text.replace('\r\n','').replace('\n','')
        location = listings.find('div',class_="location").text.replace('\r\n','').replace('\n','')
        
        info=[price,area,location]
        thewriter.writerow(info)
```
